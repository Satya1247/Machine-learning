```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6]
likes=[123,145,271,657,980,890]
plt.plot(days,likes)
```

```python
plt.title("social media analysis")
plt.xlabel("no of days")
plt.ylabel("no of likes")
plt.plot(days,likes,color="red",marker="o",markerfacecolor="black",linestyle="-.")
```

```python
fb=likes
insta=[111,234,267,678,432,567]
utube=[780,321,456,345,189,567]
```

```python
plt.plot(days,fb)
```

```python
plt.title("social media comparision")
plt.xlabel("days")
plt.ylabel("no of likes")
plt.plot(days,fb)
plt.plot(days,insta)
plt.plot(days,utube)
```

```python
plt.title("social media comparision")
plt.xlabel("days")
plt.ylabel("no of likes")
plt.plot(days,fb,color="blue",label="fb",marker="o")
plt.plot(days,insta,color="cyan",label="insta",marker="^")
plt.plot(days,utube,color="red",label="utube",marker="*")
plt.legend(loc="lower right")
plt.grid(color="orange",linewidth=2)
plt.xlim(3,5) ###
plt.ylim(400,1000)
plt.savefig("graphics1.png")
```
