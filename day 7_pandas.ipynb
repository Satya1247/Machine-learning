```python
#merge multiple datasets
```

```python
df1=pd.DataFrame({
    "name":["aa","b",'c','d'],
    "age":[12,13,14,15],
    "sub":["m",'s','c','h']
})
```

```python
df2=pd.DataFrame({
    "name":["aa","f",'g','c'],
    "age":[1,2,3,4],
    "sub":["c",'u','f','l']
})
```

```python
pd.merge(df1,df2,on="name",how="outer")
#it will take common vales
```

```python
pd.merge(df1,df2,on="name",how="inner") #default inner
#it will take common vales
```

```python
pd.merge(df1,df2,on="name",how="left") #common and dataset
#it will take common vales
```

```python
pd.merge(df1,df2,on="name",how="right") #common and second dataset
#it will take common vales
```

```python
#concatenation
pd.concat([df1,df2],ignore_index=True)
```
